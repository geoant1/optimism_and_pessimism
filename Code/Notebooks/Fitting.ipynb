```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
folder = '/Users/GA/Documents/Dayan_lab/Optimism_And_Pessimism_In_Optimised_Replay/Data/new_new_fits'
```

```python
plt.figure(figsize=(8, 5),dpi=100)

subs = [2, 8, 16, 3]
c = 1
for sub in subs:
    z    = pd.read_csv(os.path.join(folder, 'save_params_%u'%sub, 'backup.txt'), sep='\t')
    
    plt.subplot(2, 2, c)
    if c == 2 or c == 4:
        plt.ylabel(r'$\phi^{MF}$', fontsize=14)
        vals = (1-z[' param#9 '].values)
    else:
        plt.ylabel(r'$\phi^{MB}$', fontsize=14)
        vals = (1-z[' param#8 '].values)
    
    plt.scatter(range(len(vals)), vals)
    plt.ylim(-0.1, 1.1)
    plt.xlabel('Fitting iteration', fontsize=14)
    c += 1
plt.tight_layout()

plt.savefig('/Users/GA/Documents/Dayan_lab/Optimism_And_Pessimism_In_Optimised_Replay/Paper/PLOSCB/reviews/params.png')
```

```python
stds = np.logspace(np.log10(0.5), np.log10(0.02), 55)
```

```python
xs   = np.arange(55)
ys   = np.ones(55)*0.5
```

```python
plt.figure(figsize=(5, 3),dpi=100)
plt.fill_between(xs, ys+stds, ys-stds, alpha=0.5)
plt.ylabel('Parameter value', fontsize=14)
plt.xlabel('Fitting iteration', fontsize=14)
plt.tight_layout()
plt.savefig('/Users/GA/Documents/Dayan_lab/Optimism_And_Pessimism_In_Optimised_Replay/Paper/PLOSCB/reviews/variance.png')
```
